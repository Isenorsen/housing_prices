```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
%matplotlib inline
```

```python
train_data = pd.read_csv('train.csv')
test_data = pd.read_csv('test.csv')
```

```python
train_data.head()
```

## Here we can see some correlations with SalePrice values

```python
train_corr = train_data.corr()
train_corr.SalePrice.sort_values(ascending=False)
```

## This heatmap helps me to find out where is NaN values and what i need to do with them

```python
plt.figure(figsize=(25, 20))
sns.set(font_scale=1.0)
sns.heatmap(train_data.isna(), cbar=False)
```

### So, LotFrontage seems like an important column but it has too much empty values. At first i thought that it can be correlated with BldgType (maybe some types of dwelling dont have "Linear feet of street connected to property"), but by the code below i was find out that its not.

```python
nan_lotFrontage = train_data.loc[np.isnan(train_data["LotFrontage"])]
nan_lotFrontage.BldgType.value_counts()
```

```python
not_nan_lotFrontage = train_data.copy()
not_nan_lotFrontage = not_nan_lotFrontage[not_nan_lotFrontage['LotFrontage'].notnull()]
not_nan_lotFrontage.BldgType.value_counts()
```

```python
from sklearn.preprocessing import Imputer

imputer = Imputer()
lot_values = imputer.fit_transform(train_data.LotFrontage.values.reshape(-1, 1))
```

```python
train_data['LotFrontage'] = lot_values
```

```python
train_data = train_data.drop('Alley', axis=1)
train_data = train_data.drop('FireplaceQu', axis=1)
train_data = train_data.drop('PoolQC', axis=1)
train_data = train_data.drop('Fence', axis=1)
train_data = train_data.drop('MiscFeature', axis=1)
train_data = train_data.drop('Id', axis=1)

test_data = test_data.drop('Alley', axis=1)
test_data = test_data.drop('FireplaceQu', axis=1)
test_data = test_data.drop('PoolQC', axis=1)
test_data = test_data.drop('Fence', axis=1)
test_data = test_data.drop('MiscFeature', axis=1)
test_data = test_data.drop('Id', axis=1)
```

```python
plt.figure(figsize=(25, 20))
sns.heatmap(train_data.isna(), cbar=False)
```

### Many columns has "object" type, so i need transform it into "int" or "float". For automating this process i have written this function below.

```python
def factorize_objects(x):
    for item in x.columns:
        if x[item].dtype == 'O':
            x[item] = pd.factorize(x[item])[0]
            print('done')
```

```python
factorize_objects(train_data)
```

```python
garage_yr_imputer = imputer.fit_transform(train_data.GarageYrBlt.values.reshape(-1, 1))
train_data.GarageYrBlt = garage_yr_imputer
```

```python
mass_vnr_area_imputer = imputer.fit_transform(train_data.MasVnrArea.values.reshape(-1, 1))
train_data.MasVnrArea = mass_vnr_area_imputer
```

### On this stage i have dataset without NaN values and 'object' columns. 

```python
sns.heatmap(train_data.isna(), cbar=False)
```

```python
from sklearn.model_selection import train_test_split

attribs = ['MSSubClass', 'MSZoning', 'LotFrontage', 'LotArea', 'Street',
       'LotShape', 'LandContour', 'Utilities', 'LotConfig', 'LandSlope',
       'Neighborhood', 'Condition1', 'Condition2', 'BldgType', 'HouseStyle',
       'OverallQual', 'OverallCond', 'YearBuilt', 'YearRemodAdd', 'RoofStyle',
       'RoofMatl', 'Exterior1st', 'Exterior2nd', 'MasVnrType', 'MasVnrArea',
       'ExterQual', 'ExterCond', 'Foundation', 'BsmtQual', 'BsmtCond',
       'BsmtExposure', 'BsmtFinType1', 'BsmtFinSF1', 'BsmtFinType2',
       'BsmtFinSF2', 'BsmtUnfSF', 'TotalBsmtSF', 'Heating', 'HeatingQC',
       'CentralAir', 'Electrical', '1stFlrSF', '2ndFlrSF', 'LowQualFinSF',
       'GrLivArea', 'BsmtFullBath', 'BsmtHalfBath', 'FullBath', 'HalfBath',
       'BedroomAbvGr', 'KitchenAbvGr', 'KitchenQual', 'TotRmsAbvGrd',
       'Functional', 'Fireplaces', 'GarageType', 'GarageYrBlt', 'GarageFinish',
       'GarageCars', 'GarageArea', 'GarageQual', 'GarageCond', 'PavedDrive',
       'WoodDeckSF', 'OpenPorchSF', 'EnclosedPorch', '3SsnPorch',
       'ScreenPorch', 'PoolArea', 'MiscVal', 'MoSold', 'YrSold', 'SaleType',
       'SaleCondition']

train_labels = train_data.SalePrice
train_data = train_data[attribs]
```

### Now its time to scale data and choose the model.

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler

preprocess_pipeline = Pipeline([
    ('imputer', Imputer()),
    ('std_scaler', StandardScaler())
])
```

```python
scaled_data = preprocess_pipeline.fit_transform(train_data)
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(scaled_data, train_labels)
```

### Seems like GradientBoostingRegressor is the best. I tried to give it better params by GridSearchCV but failed and deleted this part of code.

```python
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor, AdaBoostRegressor
from sklearn.metrics import mean_squared_error

for_reg = RandomForestRegressor()
for_reg.fit(x_train, y_train)
for_reg_predictions = for_reg.predict(x_test)
mse_for_reg = mean_squared_error(y_test, for_reg_predictions)
rmse_for_reg = np.sqrt(mse_for_reg)
rmse_for_reg
```

```python
from sklearn.linear_model import LinearRegression

lr = LinearRegression()
lr.fit(x_train, y_train)
lr_preds = lr.predict(x_test)
mse_lr = mean_squared_error(y_test, lr_preds)
mse_lr = mean_squared_error(y_test, lr_preds)
rmse_lr = np.sqrt(mse_lr)
rmse_lr
```

```python
gbr = GradientBoostingRegressor(n_estimators=200, learning_rate=0.1, max_depth=3)
gbr.fit(x_train, y_train)
np.sqrt(mean_squared_error(y_test, gbr.predict(x_test)))
```

```python
ada = AdaBoostRegressor(base_estimator=for_reg, n_estimators=100, loss='square').fit(x_train, y_train)
np.sqrt(mean_squared_error(y_test, ada.predict(x_test)))
```
